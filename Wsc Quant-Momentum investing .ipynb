```python
Results=[]
Stocks=[]
Winners=[]
for i in range(0,302):
    d1=df2.index[0]+timedelta(bw+fw*i)
    d2=d1+timedelta(bw)
    d3=d1+timedelta(bw+fw)
    d1,d2,d3
    dfp=df2.loc[(df2.index>=d1)&(df2.index<=d2)]
    dfp2=df2.loc[(df2.index>d2)&(df2.index<=d3)]
    df_winners=df2.loc[(df2.index>d2)&(df2.index<=d3)]
   
    holdlst=[]
    for i in range(len(dfp2.columns)):
        holdlst.append((dfp2.iloc[-1,i]-dfp2.iloc[0,i])/dfp2.iloc[0,i])
        new_array = np.array(holdlst)[np.logical_not(np.isnan(np.array(holdlst)))]
    if len(new_array==0):
        t=1
    else:
        continue;
    holdreturn=sum(new_array)/(len(new_array)+t)
    
    
    
    
    lst=[]
    for i in range(len(dfp.columns)):
        lst.append((dfp.columns[i],((dfp.iloc[-1,i]-dfp.iloc[0,i])/dfp.iloc[0,i])))

    df=pd.DataFrame(lst,columns=['Stock','Returns'])
    df=df.dropna()
    df=df.sort_values('Returns',ascending=False)
    df=df.head(10)
    
    dfp2=dfp2[df['Stock'].to_list()]
    lst4=[]
    for j in range(len(dfp2.columns)):
        lst4.append((dfp2.columns[j],(dfp2.iloc[-1,j]-dfp2.iloc[0,j])/dfp2.iloc[0,j]))
    df_final=pd.DataFrame(lst4,columns=['Stock','Returns'])
    if len(df_final==0):
        t=1;
    else:
        continue;
    df_final['Returns'].sum()/len(df_final)
    
    Results.append((d2,holdreturn,df_final['Returns'].sum()/len(df_final)))
    Stocks.append((df['Stock']))
```
